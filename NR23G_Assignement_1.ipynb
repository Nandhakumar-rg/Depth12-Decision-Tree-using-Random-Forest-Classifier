```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
import pandas as pd
```

```python
# Load data
train_data = np.loadtxt('madelon_train.data')
train_labels = np.loadtxt('madelon_train.labels')

valid_data = np.loadtxt('madelon_valid.data')
valid_labels = np.loadtxt('madelon_valid.labels')
```

```python
# Values of k
k_values = [3, 10, 30, 100, 300]

# Initialize arrays to store errors
train_errors = []
test_errors = []
```

```python
# Train random forests
for k in k_values:
    rf_model = RandomForestClassifier(n_estimators=k, max_features=int(np.sqrt(500)), random_state=42)
    rf_model.fit(train_data, train_labels)
    train_preds = rf_model.predict(train_data)
    valid_preds = rf_model.predict(valid_data)
    train_error = 1 - accuracy_score(train_labels, train_preds)
    test_error = 1 - accuracy_score(valid_labels, valid_preds)
    train_errors.append(train_error)
    test_errors.append(test_error)
```

```python
# Plot the errors
plt.plot(k_values, train_errors, marker='o', linestyle='-', label='Training Error')
plt.plot(k_values, test_errors, marker='o', linestyle='-', label='Test Error')
plt.xlabel('Number of Trees (k)')
plt.ylabel('Misclassification Error')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Report misclassification errors in a table
error_table = {'k': k_values, 'Training Error': train_errors, 'Test Error': test_errors}
print("{:<5} {:<15} {:<15}".format('k', 'Training Error', 'Test Error'))
for i in range(len(k_values)):
    print("{:<5} {:<15.5f} {:<15.5f}".format(k_values[i], train_errors[i], test_errors[i]))
```
